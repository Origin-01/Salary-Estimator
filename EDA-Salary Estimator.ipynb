```python
# The categorical columns of the data
df_category = df[['Job Title','Rating','Location','Headquarters','Size','Founded','Industry','Sector','Revenue','avg_sal', 'python', 'r studio',
       'excel', 'spark', 'city', 'state', 'hq_city', 'job_simp', 'job_title',
       'num_comp']]
df_category
```

```python
# Histograms of various columns 
import matplotlib.pyplot as plt
import seaborn as sns

for i in df_category.columns:
    col=df_category[i].value_counts()
    print('Graph of %s :Total = %d' % (i,len(col)))# to name the graphs and show the total number of categories under each heading
    graphs=sns.barplot(x=col.index,y=col)
    graphs.set_xticklabels(graphs.get_xticklabels(),rotation=90)# to rotate the labels of the x axis by 90
    plt.show()
```

```python
# Some of the graph shown below does not explain the barplot hence we take 20 rows of each section

for i in df_category[['Job Title','Location','Headquarters','Founded','Industry','city','state','hq_city']].columns:
    col=df_category[i].value_counts()[:20]
    print('Graph of %s :Total = %d' % (i,len(col)))# to name the graphs and show the total number of categories under each heading
    graphs=sns.barplot(x=col.index,y=col)
    graphs.set_xticklabels(graphs.get_xticklabels(),rotation=90)# to rotate the labels of the x axis by 90
    plt.show()
```

```python
# Lastly for the job description we can make a word cloud for it

# Import the libraries
!pip install wordcloud
!pip install nltk

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from wordcloud import WordCloud, ImageColorGenerator,STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
print('Imported')
```

```python
words=' '.join(df['Job Description'])

# Removing punctuation and stop words by creating a function
def punctuation_stop(text):
    filtered=[]
    stop_words=set(stopwords.words('english'))
    word_tokens=word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered

word_filtered=punctuation_stop(words)

text=' '.join([ele for ele in word_filtered])

wc=WordCloud(background_color='white',random_state=1,stopwords=STOPWORDS,max_words=2000,width=800,height=1500)
wc.generate(text)

plt.figure(figsize=[10,10])
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
